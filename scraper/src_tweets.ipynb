```python
import pandas as pd
import twint
import nest_asyncio
nest_asyncio.apply()
```

```python
def get_tweet(keywords):
    '''
    input: file name containing keywords
    output: dataframe, list of tweets
    '''
    c = twint.Config()

    with open(keywords,'r') as f:
        df = pd.DataFrame()
        for keywords in f:
            c.Search = keywords
            c.Limit = 30
            c.Pandas = True
            twint.run.Search(c)
            data = twint.storage.panda.Tweets_df
            df = df.append(data, ignore_index=True)
        #print(df)
    
    return df 
```

```python
keywords_list = '../search-keywords/keywords-mix-150522.txt'
df = get_tweet(keywords_list)
```
